```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_final_negatives.csv')
data
```

```python
X = np.array(data.iloc[:,1:4805])
y = data.iloc[:,-1]

X_1 = X[:,:3776]
X_2 = X[:,3780:]
X = np.hstack([X_1,X_2])

#Due to some mistake,the final negatives class was stored as label 44 instead of 43
#Correcting the error
y[y == 44] = 43

y = np.array(y).reshape([-1,1])

# print(X[:,-1])
# print(y)
# print(X)
num_classes = int(np.max(y))
train_t = None
val_t = None
train_y = None
val_y = None

print(num_classes)
for i in range(1,num_classes + 1):
    # print(i)
    indices = np.where(y == i)
    # print(indices)
    train_id = indices[0][0] + int(0.8*len(indices[0]))
    # print(indices)
    # print(train_id)
    x_train = X[indices[0][0]:train_id,:]
    x_val = X[train_id:indices[0][-1],:]
    y_train = (y[indices[0][0]:train_id,0]).reshape([-1,1])
    y_val = (y[train_id:indices[0][-1],0]).reshape([-1,1])
    
#     print(x_train.shape)
#     print(x_val.shape)
    if i == 1:
        train_t = x_train
        val_t = x_val
        train_y = y_train
        val_y = y_val
    else:
        train_t = np.vstack((train_t,x_train))
        val_t = np.vstack((val_t,x_val))
        train_y = np.vstack((train_y,y_train))
        val_y = np.vstack((val_y,y_val))

print(val_t.shape)
print(train_t.shape)
print(val_y.shape)
print(train_y.shape)

X_train,y_train,X_val,y_val = train_t,train_y,val_t,val_y

# print(num_classes)
min_in_y = np.min(y)
# print(num_classes)
# print(min_in_y)
# print(y[2069,0])
# print(y.shape)
# print(X.shape)
y_one_hot = np.zeros((X.shape[0],num_classes))
y_one_hot_train = np.zeros((X_train.shape[0],num_classes))
y_one_hot_val = np.zeros((X_val.shape[0],num_classes))

# print(y_one_hot.shape)

for i in range(y_train.shape[0]):
    y_one_hot_train[i,int(y_train[i,0])-1] = 1
for i in range(y_val.shape[0]):
    y_one_hot_val[i,int(y_val[i,0])-1] = 1
```

```python
WIDTH = 64
num_features = int(X.shape[1])
HEIGHT = int(num_features/WIDTH)

X_train = X_train.reshape([-1,WIDTH,HEIGHT,1])
X_val = X_val.reshape([-1,WIDTH,HEIGHT,1])
print(X_train.shape)
print(X_val.shape)
```

```python
def net(x):
    #64*75*1
    conv1 = tf.nn.dropout(tf.layers.conv2d(x,64,kernel_size=[4,4],strides=[1,1],padding='same'),keep_prob=0.5)
    a1 = tf.nn.relu(conv1)
    pool1 = tf.layers.max_pooling2d(a1,pool_size=[2,2],strides=[2,2])

    #32*37*64
    conv2 = tf.nn.dropout(tf.layers.conv2d(pool1,128,kernel_size=[4,4],strides=[1,1],padding='same'),keep_prob=0.5)
    a2 = tf.nn.relu(conv2)
    pool2 = tf.layers.max_pooling2d(a2,pool_size=[2,2],strides=[2,2])
    
#     #16*18*128
#     conv3 = tf.nn.dropout(tf.layers.conv2d(pool2,256,kernel_size=[4,4],strides=[1,1],padding='same'),keep_prob=0.5)
#     a3 = tf.nn.relu(conv3)
#     pool3 = tf.layers.max_pooling2d(a3,pool_size=[2,2],strides=[2,2])
    
    #16*18*128 - flattened
    flat3 = tf.contrib.layers.flatten(pool2)
    full3 = tf.layers.dense(flat3,1024)
    a3 = tf.nn.relu(full3)
    
#     full5 = tf.layers.dense(a4,1024)
#     a5 = tf.nn.relu(full5)
    
    full4 = tf.layers.dense(a3,num_classes)
    probs = tf.nn.sigmoid(full4)
    
    return probs

#
# tf.reset_default_graph()
# probs = net(tf.ones([50,64,75,1]))
# print(probs.shape)
```

```python
tf.reset_default_graph()
num_epochs = 10
lr = 1e-4
batch_size = 32

sess = tf.Session()

#Placeholders
X_tf = tf.placeholder(shape=[None,WIDTH,HEIGHT,1],name='X_tf',dtype=tf.float32)
y_tf = tf.placeholder(shape=[None,num_classes],name='y_tf',dtype=tf.float32)

probs = net(X_tf)
y_hat = tf.argmax(probs,axis=1)
y_compare = tf.argmax(y_tf,axis=1)
accuracy_ = tf.reduce_mean(tf.cast(tf.equal(y_hat,y_compare),tf.float32))

loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=probs,labels=y_tf))

trainer = tf.train.AdamOptimizer(lr)
train_step = trainer.minimize(loss)

sess.run(tf.global_variables_initializer())
sess.run(tf.local_variables_initializer())

loss_vector = []
train_acc_vector = []
val_acc_vector = []

for epoch in range(num_epochs):
    print('Epoch ' + str(epoch) + ' chal rha hai abhi')
    
    perms = np.random.permutation(X_train.shape[0])
    num_batches = (int)(X_train.shape[0]/batch_size)

    X_shuffle = X_train[perms,:]
    y_shuffle = y_one_hot_train[perms,:]
    
    epoch_loss = 0.0
    epoch_acc = 0.0
    epoch_val_acc = 0.0
    
    lr = 1.0
    for i in range(num_batches):
        print(i)
        X_batch = X_shuffle[i*batch_size:(i+1)*batch_size,:]
        y_batch = y_shuffle[i*batch_size:(i+1)*batch_size,:]
        
        _,l,acc = sess.run([train_step,loss,accuracy_],feed_dict={X_tf:X_batch,y_tf:y_batch})
        # print('it : ' + str(it))
        epoch_loss+=l
        epoch_acc+=acc
    
    perms = np.random.permutation(X_val.shape[0])

    X_val_shuffle = X_val[perms,:]
    y_val_shuffle = y_one_hot_val[perms,:]
    
    X_val_batch = X_val_shuffle[0:batch_size,:]
    y_val_batch = y_val_shuffle[0:batch_size,:]
    
    val_acc_list = sess.run([accuracy_],feed_dict={X_tf:X_val_batch,y_tf:y_val_batch})    
    val_acc = val_acc_list
    # print('it : ' + str(it))
    
    print('Learning Rate chal rha hai : ' + str(lr))
    loss_vector.append(epoch_loss/(num_batches*1.0))
    train_acc_vector.append(epoch_acc/(num_batches*1.0))
    val_acc_vector.append(val_acc)
    
    if epoch%1 == 0:
        print('Epoch = '+ str(epoch))
        print('Loss = ' + str(epoch_loss/(num_batches*1.0)))
        print('Accuracy = ' + str(epoch_acc/(num_batches*1.0)))
        print('Validation Accuracy = ' + str(val_acc))
```

```python
print('Final-Loss : ' + str(loss_vector[-1]))
print('Final-Train-Accuracy : ' + str(train_acc_vector[-1]))
print('Final-Val-Accuracy : ' + str(val_acc_vector[-1]))
```

```python
epochs = [i for i in range(num_epochs)]
plt.plot(epochs,loss_vector)
plt.figure()
plt.plot(epochs,val_acc_vector)
plt.plot(epochs,train_acc_vector)
plt.show()
```
